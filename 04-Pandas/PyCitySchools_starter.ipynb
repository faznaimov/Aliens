```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
student_data.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#unique schools
total_schools = school_data_complete["School ID"].nunique()

#unique students
total_students = school_data_complete["Student ID"].nunique()

#total budget
total_budget = school_data["budget"].sum()

#average math score
avg_math = school_data_complete["math_score"].mean()

#average reading score
avg_reading = school_data_complete["reading_score"].mean()

#calculation of percentage passing math
pass_math = school_data_complete[school_data_complete["math_score"]>=70]
pass_math_prc = (pass_math["math_score"].count()/total_students)*100

#calculation of percentage passing reading
pass_reading = school_data_complete[school_data_complete["reading_score"]>=70]
pass_reading_prc = (pass_reading["reading_score"].count()/total_students)*100

#overall passing rate
total_pass = (avg_math + avg_reading)/2

#created dataframe with all above data
d_summary = pd.DataFrame({
    "Total Schools":[total_schools],
    "Total Students":[total_students],
    "Total Budget":[total_budget],
    "Average Math Score":[avg_math],
    "Average Reading Score":[avg_reading],
    "% Passing Math":[pass_math_prc],
    "% Passing Reading":[pass_reading_prc],
    "% Overall Passing Rate":[total_pass]})

#formatting
d_summary["Total Students"]=d_summary["Total Students"].map("{:,}".format)
d_summary["Total Budget"]=d_summary["Total Budget"].map("${:,.2f}".format)

d_summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#groups by school
by_school = school_data_complete.set_index('school_name').groupby(['school_name'])

#school types
types = school_data.set_index('school_name')['type']

# total students by school
students = by_school['Student ID'].count()

# school budget
budget = school_data.set_index('school_name')['budget']

#per student budget
stu_budget = budget/students

#avg scores by school
avg_math = by_school['math_score'].mean()
avg_read = by_school['reading_score'].mean()

# % passing scores
pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()/students
pass_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/students
overall = (pass_math+pass_read)/2

#creating a dataframe
sch_summary = pd.DataFrame({
    "School Type": types,
    "Total Students": students,
    "Per Student Budget": stu_budget,
    "Total School Budget": budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "% Passing Math": pass_math,
    "% Passing Reading": pass_read,
    "Overall Passing Rate": overall})

#formatting
sch_summary["Total Students"]=sch_summary["Total Students"].map("{:,}".format)
sch_summary["Total School Budget"]=sch_summary["Total School Budget"].map("${:,.2f}".format)
sch_summary["Per Student Budget"]=sch_summary["Per Student Budget"].map("${:.2f}".format)
sch_summary["Average Math Score"]=sch_summary["Average Math Score"].map("{:.2f}".format)
sch_summary["Average Reading Score"]=sch_summary["Average Reading Score"].map("{:.2f}".format)
sch_summary["% Passing Math"]=sch_summary["% Passing Math"].map("{:.3%}".format)
sch_summary["% Passing Reading"]=sch_summary["% Passing Reading"].map("{:.3%}".format)
sch_summary["Overall Passing Rate"]=sch_summary["Overall Passing Rate"].map("{:.3%}".format)

top_schools=sch_summary.sort_values("Overall Passing Rate", ascending=False)
top_schools.head(5)
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
worst_schools=sch_summary.sort_values("Overall Passing Rate", ascending=True)
worst_schools.head(5)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#extracting data from each year that is grouped by school and calculating mean
nine=student_data.loc[student_data["grade"]=="9th"].groupby("school_name")["math_score"].mean()
ten=student_data.loc[student_data["grade"]=="10th"].groupby("school_name")["math_score"].mean()
eleven=student_data.loc[student_data["grade"]=="11th"].groupby("school_name")["math_score"].mean()
twelve=student_data.loc[student_data["grade"]=="12th"].groupby("school_name")["math_score"].mean()

#creating dataframe with above info
math_scores=pd.DataFrame({"9th Grade":nine,
                         "10th Grade":ten,
                         "11th Grade":eleven,
                         "12th Grade":twelve})
#formatting
math_scores["9th Grade"]=math_scores["9th Grade"].map("{:.2f}%".format)
math_scores["10th Grade"]=math_scores["10th Grade"].map("{:.2f}%".format)
math_scores["11th Grade"]=math_scores["11th Grade"].map("{:.2f}%".format)
math_scores["12th Grade"]=math_scores["12th Grade"].map("{:.2f}%".format)
math_scores
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
#creating new DF
by_spending=pd.DataFrame({
    "Per Student Budget":stu_budget,
    "Average Math Score":avg_math,
    "Average Reading Score":avg_read,
    "% Passing Math":pass_math,
    "% Passing Reading":pass_read,
    "Overall Passing Rate":overall
})

#breaking down to bins
by_spending["Spending Ranges (Per Student)"]=pd.cut(by_spending["Per Student Budget"],spending_bins,labels=group_names)

#grouping by new bin column
by_spending=by_spending.groupby("Spending Ranges (Per Student)")["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","Overall Passing Rate"].mean()

#formatting
by_spending["% Passing Math"]=by_spending["% Passing Math"].map("{:.2%}".format)
by_spending["% Passing Reading"]=by_spending["% Passing Reading"].map("{:.2%}".format)
by_spending["Overall Passing Rate"]=by_spending["Overall Passing Rate"].map("{:.2%}".format)

by_spending
```
